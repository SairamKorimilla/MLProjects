### Question 2:  Using MLP to simulate the 10x10 multiplication table:

####  You need to build a MLP model for  $   {y = x_1 * x_2} $:
   - <span class="mark">Start from Step 3, then complete the Step 4 and Step 5.</span>
     




### Step 1.  Generate a DataFrame for 10x10 data  


```python
import pandas as pd

# if n= 9  for data 9x9 table
n = 10
x1 = [i   for i in range(1, n+1) for j in range(1, n+1) ]
x2 = [j   for i in range(1, n+1) for j in range(1, n+1) ]
y =  [i*j for i in range(1, n+1) for j in range(1, n+1)  ]

data =pd.DataFrame({"x1":x1, "x2":x2, "y":y})
data.head()
```

### Step 2. Prepare data for Machine Learning

```python
# Splitting the dataset into  training and testing sets
from sklearn.model_selection import train_test_split
training_set, test_set = train_test_split(data, test_size = 0.2, random_state = 21)

# trainning datasets
X_train = training_set.iloc[:, 0:-1].values
y_train = training_set.iloc[:,-1].values

# test
X_test = test_set.iloc[:,0:-1].values
y_test = test_set.iloc[:,-1].values

# display:
for row in range(len(X_test)):
    print(f'{X_test[row][0]:2} x {X_test[row][1]:2} => {y_test[row]:2}')
```

# HOMEWORK
### Step 3. Train a MLP ( or ANN ) using SK-learn MLPRegressor()

https://scikit-learn.org/dev/modules/generated/sklearn.neural_network.MLPRegressor.html

    - try different parameter settings:
        0. activation function: {‘identity’, ‘logistic’, ‘tanh’, ‘relu’}, default=’relu’   
        1. hidden_layer_sizes:  {100, 200, 300, ..... }, default=(100,)
        2. solver: {‘lbfgs’, ‘sgd’, ‘adam’}, default=’adam’
        3. learning_rate_init: {0.001, 0.01, 0.1, 0.2}, default = 0.001,
        4. max_iter: {1000, 2000, 3000, ...}, default = 200
        5. random_state: { None, 1, 100}, default = None

    
    - Use different parameter combinations to train your model with the given training data

    - Try to get a model with 100% accuracy (using round y_hat) on test data
    
    - Write down your best model.

   

```python
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import accuracy_score,mean_squared_error
from sklearn.preprocessing import MinMaxScaler
import numpy as np


# You need to reset the regressor: the default is not good enough
# mlp = MLPRegressor() # default
# Normalize the data using MinMaxScaler
scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Initialize the MLPRegressor with a simplified architecture
mlp = MLPRegressor(
    hidden_layer_sizes=(120, 60, 30),    # Simplified: Three hidden layers with fewer neurons
    activation='relu',                # Using ReLU as activation function
    solver='lbfgs',                    # lbfgs solver
    learning_rate_init=0.00005,        # Lower learning rate for fine-grained learning
    max_iter=25000,                   # More iterations
    random_state=1,                   # Random state for reproducibility
)
```

```python
#Fitting the training data to the network
mlp.fit(X_train_scaled, y_train)
```

## Step 4. Test the testing samples and print out the MSE
   


```python
#Predict y_pred values for X_test
y_pred = mlp.predict(X_test_scaled)


# print the y_pred, y_hat (round to int), y_test, and MSE
y_hat = np.round(y_pred).astype(int)
for i in range(len(X_test)):
    print(f"{'y_pred:'}{y_pred[i]:>6.2f}         {'y_hat:'}{y_hat[i]:>3}       {'y_test:'}{y_test[i]:>3}")

# find and print the MSE
mse_before_rounding = mean_squared_error(y_test, y_pred)
print(f"\nMSE before rounding: {mse_before_rounding:.4f}")
```

```python
from sklearn.metrics import accuracy_score
# use y_hat (the y_pred rounded)  and  y_test(real y value) to calculate MSE and the accuracy.

# MSE
mse_after_rounding = mean_squared_error(y_test, y_hat)
print(f"The MSE after rounding: {mse_after_rounding:.1f}")


# accuracy
accuracy = np.mean(y_hat == y_test) * 100
print(f"The accuracy: {accuracy:.1f}%")
print('y_hat: ',list(y_hat))
print('y_test:',list(y_test))

```

## Step 5. Test new data

    Given new samples: x1 = 11:   
     - x2 = 1   
     - x2 = 2   
     - x2 = 11  

```python
# new samples: x1 = 11, x2 = 1 or 2 or 11

new_sample = [[11,  1],
              [11,  2],
              [11,  11]]

# Scale the new samples using the same scaler used for training and testing data
new_samples_scaled = scaler.transform(new_sample)

# Predict outcomes using the trained MLP model
new_predictions = mlp.predict(new_samples_scaled)

# Round the predictions to get integer results
new_predictions_rounded = np.round(new_predictions).astype(int)

print('The predicted products:',new_predictions_rounded)



```

### The last model:

  
  
  
  




This MLP model configuration employs three hidden layers with 120, 60, and 30 neurons, balancing complexity and capacity to learn patterns while avoiding overfitting. The ReLU activation function introduces necessary non-linearity and accelerates training efficiency. The ‘lbfgs’ solver is optimal for small datasets, ensuring quick convergence and memory efficiency. A low learning rate of 0.00005 allows for fine-tuned weight adjustments, preventing divergence during training. With a maximum of 25,000 iterations, the model has ample time to minimize the loss function. Setting a random state ensures reproducibility of results across different runs.
